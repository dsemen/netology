```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
s = np.random.poisson(5, 1000)
count, bins, ignored = plt.hist(s, 14, normed=True,label='Пуассон гистограмма' )

plt.legend(loc='upper right')
```

```python
norm_rv = sts.norm(2, 0.5)
norm_rv2 = sts.norm(2, 1)
norm_rv3 = sts.norm(0.5, 0.5)
x = np.linspace(0,4,1000)
pdf = norm_rv.pdf(x)
pdf2 = norm_rv2.pdf(x)
pdf3 = norm_rv3.pdf(x)
cdf = norm_rv.cdf(x)
cdf2 = norm_rv2.cdf(x)
cdf3 = norm_rv3.cdf(x)
plt.figure(figsize=(15,8))
plt.plot(x, pdf)
plt.plot(x, pdf2)
plt.plot(x, pdf3)
plt.plot(x, cdf)
plt.plot(x, cdf2)
plt.plot(x, cdf3)

plt.legend(['pdf 2:0.5',  'pdf 2:1', 'pdf 0.5:0.5','cdf 2:0.5',  'cdf 2:1', 'cdf 0.5:0.5'])
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

```python
random_val = sts.expon(3)
random_val1 = sts.expon(6)
x = np.linspace (0,10,100)
cdf = random_val.cdf(x)
cdf1 = random_val1.cdf(x)
plt.plot(x, cdf, label='Экспонента')
plt.plot(x, cdf1, label='Экспонента2')
```

```python
import scipy.special as sps
shape, scale = 2., 2.
s1 = np.random.gamma(shape, scale, 1000)
count, bins, ignored = plt.hist(s1, 50, normed=True)
y = bins**(shape-1)*(np.exp(-bins/scale) /
                     (sps.gamma(shape)*scale**shape))
plt.plot(bins, y, linewidth=2, color='r',label='Гамма распределение,которое не изучали' )

plt.legend(loc='upper right')
```
