```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
pois_rv = sts.poisson(1)
pois_rv2 = sts.poisson(0.5)
pois_rv3 = sts.poisson(3)
x = np.linspace(0,20,21)
cdf = pois_rv.cdf(x)
cdf2 = pois_rv2.cdf(x)
cdf3 = pois_rv3.cdf(x)
plt.figure(figsize=(10,10))
plt.plot(cdf)
plt.plot(cdf2)
plt.plot(cdf3)
plt.title('Распределение Пуассона cdf')
plt.legend(['cdf 1',  'cdf 0.5', 'cdf 3'])
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

```python
plt.figure(figsize=(10,10))
pmf = pois_rv.pmf(x)
pmf2 = pois_rv2.pmf(x)
pmf3 = pois_rv3.pmf(x)
plt.plot(pmf)
plt.plot(pmf2)
plt.plot(pmf3)
plt.title('Распределение Пуассона pmf')
plt.legend(['pmf 1',  'pmf 0.5', 'pmf 3'])
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

```python
random_val = sts.expon(3)
random_val1 = sts.expon(5)
x = np.linspace (5,10,100)
cdf = random_val.cdf(x)
cdf1 = random_val1.cdf(x)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.plot(x, cdf)
plt.plot(x, cdf1)
plt.legend(['expon pdf 3','expon pdf 5'])
plt.title('Экспоненциальное распределение')
```

```python
cauchy_rv = sts.cauchy()

x = np.linspace(0,2,100)
y = np.linspace(0,5,100)
z = np.linspace(0,7, 100)
pdf = cauchy_rv.pdf(x)
pdf2 = cauchy_rv.pdf(y)
pdf3 = cauchy_rv.pdf(z)

plt.plot(pdf)
plt.plot(pdf2)
plt.plot(pdf3)

plt.title('Распределение Коши')
plt.legend(['pdf 0:2',  'pdf 0:5', 'pdf 0:7'])
```

```python
cdf = cauchy_rv.cdf(x)
cdf2 = cauchy_rv.cdf(y)
cdf3 = cauchy_rv.cdf(z)
plt.figure(figsize=(10,10))
plt.plot(cdf)
plt.plot(cdf2)
plt.plot(cdf3)
plt.title('Распределение Коши')
plt.legend(['cdf 0:2', 'сdf 0:5', 'cdf 0:7'])
```
