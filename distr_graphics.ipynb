```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
s = np.random.poisson(5, 1000)
count, bins, ignored = plt.hist(s, 14, normed=True,label='Пуассон гистограмма' )

plt.legend(loc='upper right')
```

```python
N = 30
y = np.random.poisson(5, 30)
x1 = np.linspace(0, 10, N, endpoint=True)

plt.plot(x1, y, 'o',label='Пуассон точки')
plt.legend(loc='upper left')
plt.show()
```

```python
random_val = sts.expon(3)
random_val1 = sts.expon(6)
x = np.linspace (0,10,100)
cdf = random_val.cdf(x)
cdf1 = random_val1.cdf(x)
plt.plot(x, cdf, label='Экспонента')
plt.plot(x, cdf1, label='Экспонента2')
```

```python
import scipy.special as sps
shape, scale = 2., 2.
s1 = np.random.gamma(shape, scale, 1000)
count, bins, ignored = plt.hist(s1, 50, normed=True)
y = bins**(shape-1)*(np.exp(-bins/scale) /
                     (sps.gamma(shape)*scale**shape))
plt.plot(bins, y, linewidth=2, color='r',label='Гамма распределение,которое не изучали' )

plt.legend(loc='upper right')
```
