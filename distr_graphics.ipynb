```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
from scipy.stats import poisson
%matplotlib inline
```

```python
pois_rv = sts.poisson(1)
pois_rv2 = sts.poisson(0.5)
pois_rv3 = sts.poisson(3)
x = np.linspace(0,20,21)
cdf = pois_rv.cdf(x)
cdf2 = pois_rv2.cdf(x)
cdf3 = pois_rv3.cdf(x)
plt.figure(figsize=(10,10))
plt.plot(cdf)
plt.plot(cdf2)
plt.plot(cdf3)
plt.title('Распределение Пуассона')
plt.legend(['1',  '0.5', ' 3'])
plt.ylabel('$cdf(x)$')
plt.xlabel('$x$')
```

```python
x = np.linspace(0,50,51)
for N in [5, 10, 13, 20]:
    
    rv = poisson(N)
    pmf = rv.pmf(x)
    plt.plot(x, pmf, 'o', label="$N=%s" % (N))
plt.legend()
plt.title("Poisson scatter")

plt.ylabel('$PMF(X)$')
plt.xlabel('$x$')
```

```python
expon_rv = sts.expon(1)
expon_rv2 = sts.expon(2)
expon_rv3 = sts.expon(3)
expon_rv4 = sts.expon(4)
x = np.linspace(3,20,21)
cdf = expon_rv.cdf(x) 
cdf2 = expon_rv2.cdf(x) 
cdf3 = expon_rv3.cdf(x) 
cdf4 = expon_rv4.cdf(x) 
plt.figure(figsize=(15,8))
plt.plot(x, cdf)
plt.plot(x, cdf2)
plt.plot(x, cdf3)
plt.plot(x, cdf4)
plt.title("График экспоненциальной функции распределения")
plt.legend(['lambda=1',  'lambda=2', 'lambda=3', 'lambda=4'])
plt.ylabel('$CDF(x)$')
plt.xlabel('$x$')
```

```python
expon_rv = sts.expon(1)
expon_rv2 = sts.expon(2)
expon_rv3 = sts.expon(3)
expon_rv4 = sts.expon(4)
x = np.linspace(5,20,30)
pdf = expon_rv.pdf(x)
pdf2 = expon_rv2.pdf(x)
pdf3 = expon_rv3.pdf(x)
pdf4 = expon_rv4.pdf(x)
plt.figure(figsize=(15,8))
plt.plot(x, pdf)
plt.plot(x, pdf2)
plt.plot(x, pdf3)
plt.plot(x, pdf4)
plt.title("График плотности экспоненциального распределения")
plt.legend(['lambda=1',  'lambda=2', 'lambda=3', 'lambda=4'])
plt.ylabel('$PDF(x)$')
plt.xlabel('$x$')
```

```python
cauchy_rv = sts.cauchy()

x = np.linspace(0,2,100)
y = np.linspace(0,5,100)
z = np.linspace(0,7, 100)
pdf = cauchy_rv.pdf(x)
pdf2 = cauchy_rv.pdf(y)
pdf3 = cauchy_rv.pdf(z)

plt.plot(pdf)
plt.plot(pdf2)
plt.plot(pdf3)

plt.title('Распределение Коши')
plt.legend(['0:2',  '0:5', '0:7'])
plt.ylabel('$PDF(x)$')
plt.xlabel('$x$')
```

```python
cdf = cauchy_rv.cdf(x)
cdf2 = cauchy_rv.cdf(y)
cdf3 = cauchy_rv.cdf(z)
plt.figure(figsize=(10,10))
plt.plot(cdf)
plt.plot(cdf2)
plt.plot(cdf3)
plt.title('Распределение Коши')
plt.legend(['0:2', '0:5', '0:7'])
plt.ylabel('$CDF(x)$')
plt.xlabel('$x$')
```
