```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
season_tags=pd.read_excel('season_tags.xlsx')
```

```python
plt.title("Коэффиценты сезонности продаж в штуках по направлениям ИФ/БК")
plt.plot(season_tags['Month'], season_tags['BK'], marker='o')
plt.plot(season_tags['IF'],marker='o')
plt.legend()
"""Взял этот график, чтобы сравнить динамику продаж между двумя направлениями, 
визуально можно легко увидеть, в каком месяце провал, в каком рост и по какому направлению"""
```

```python
season_tags [['IF','BK','Month']].plot(kind='bar',x='Month')
plt.xlabel('Месяц')
"""Сделал то же самое что и выше, только поменял на столбики, с понятийной точки зрения то же самое"""
```

```python
# повторяю все то же самое на плотли
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly
import plotly.graph_objs as go
init_notebook_mode(connected=True)
trace0 = go.Scatter(
    x=season_tags.Month,
    y=season_tags.BK,
    name='Изделия без камней'
)
trace1 = go.Scatter(
    x=season_tags.Month,
    y=season_tags.IF,
    name='Изделия с фианитами'
)
data = [trace0, trace1]
layout = {'title': 'Коэффиценты сезонности по направлениям'}
fig = go.Figure(data=data, layout=layout)
iplot(fig, show_link=False)
```

```python
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly
import plotly.graph_objs as go
init_notebook_mode(connected=True)
trace0 = go.Bar(
    x=season_tags.Month,
    y=season_tags.BK,
    name='Изделия без камней'
)
trace1 = go.Bar(
    x=season_tags.Month,
   y=season_tags.IF,
    name='Изделия с фианитами'
)
data = [trace0, trace1]
    
layout = {'title': 'Коэффиценты сезонности по направлениям', 'xaxis': {'title': 'Month'}}
fig = go.Figure(data=data, layout=layout)
iplot(fig, show_link=False)
```
